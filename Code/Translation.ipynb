```python
import pandas as pd
```

```python
df = pd.read_csv('merged_real_estate.csv', sep=',', encoding='utf-8-sig')
```

Selecting columns

```python
col = ['Comune_cat','Sez','Fascia','LinkZona','Cod_Tip','Stato_prev','Sup_NL_compr','Sup_NL_loc','Unnamed: 21']
df_sel = df.drop(columns=col)

df_sel.head()
```

Translating

```python
# Define all column renames
column_renames = {
    "Area_territoriale": "area",
    "Regione": "region",
    "Prov": "prov",
    "Comune_ISTAT": "mun_istat",
    "Comune_amm": "mun_cad",
    "Comune_descrizione": "mun_name",
    "Zona": "omi",
    "Descr_Tipologia": "type",
    "Stato": "state",
    "Compr_min": "buy_min",
    "Compr_max": "buy_max",
    "Loc_min": "rent_min",
    "Loc_max": "rent_max",
    "building_id": "id"
}

df_sel = df_sel.rename(columns=column_renames)

# Apply value mappings
df_sel["area"] = df_sel["area"].replace({
    "NORD-OVEST": "NW",
    "NORD-EST": "NE",
    "CENTRO": "C",
    "ISOLE": "I",
    "SUD": "S"
})

df_sel["type"] = df_sel["type"].replace({
    "Abitazioni civili": "Residential housing",
    "Box": "Garage / Box",
    "Ville e Villini": "Villas and small houses",
    "Negozi": "Shops",
    "Abitazioni di tipo economico": "Low-cost housing",
    "Magazzini": "Warehouses",
    "Uffici": "Offices",
    "Laboratori": "Laboratories",
    "Capannoni tipici": "Typical industrial buildings",
    "Capannoni industriali": "Industrial buildings",
    "Autorimesse": "Garages",
    "Posti auto scoperti": "Uncovered parking spaces",
    "Posti auto coperti": "Covered parking spaces",
    "Centri commerciali": "Shopping centers",
    "Uffici strutturati": "Structured offices",
    "Abitazioni tipiche dei luoghi": "Typical local housing",
    "Abitazioni signorili": "Luxury housing",
    "Pensioni e assimilati": "Guesthouses and similar",
    "Fabbricati e locali per esercizi sportivi": "Sports facilities"
})

df_sel["state"] = df_sel["state"].replace({
    "NORMALE": "Normal",
    "OTTIMO": "Excellent",
    "SCADENTE": "Poor"
})
```

Ordering variables

```python
new_order = [
    "id", "semester", "area", "region", "prov", "mun_istat", "mun_cad", "mun_name",
    "omi", "type", "state", "buy_min", "buy_max", "rent_min", "rent_max"
]

df_sel = df_sel[new_order]
```
